```python
import pickle
import matplotlib.pyplot as plt
from pathlib import Path
def draw(path):
    root = Path(path)
    for data in root.iterdir():
        try:
            log = pickle.load(data.open('rb'))
            plt.plot(log, label=data)
        except:
            pass
    plt.legend()
    plt.show()
```

```python
draw('.')
```
