```python
import matplotlib.pyplot as plt
```

```python
import pickle
from pathlib import Path
def draw(path):
    root = Path(path)
    for data in root.iterdir():
        log = pickle.load(data.open('rb'))
        plt.plot(log, label=data)
    plt.legend()
    plt.show()
```

```python
draw('text_log')
```

```python
draw('log/draw_valid')
```
